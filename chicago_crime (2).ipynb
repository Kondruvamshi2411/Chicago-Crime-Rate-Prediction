```python
chicago_prophet = chicago_data.resample('M').size().reset_index()
```

```python
chicago_prophet.columns =['Date','Count']
```

```python
chicago_prophet
```

```python
chicago_final_data = chicago_prophet.rename(columns={'Date':'ds','Count':'y'})
```

```python
chicago_final_data
```

```python
from prophet import Prophet
m = Prophet()
m.fit(chicago_final_data)
```

```python
future = m.make_future_dataframe(periods=730)
```

```python
prediction = m.predict(future)
```
